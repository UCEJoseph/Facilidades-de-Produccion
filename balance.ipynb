```python
# Universidad Central del Ecuador
# Facultad de Ingeniería en Geología, Minas, Petróleos y Ambiental
# Carrera de Petróleos
# Integrantes: Joseph Pacheco
# Tema: Balance Volumétrico

import pandas as pd
import math
from tabulate import tabulate
from pandas.core.groupby.groupby import T

print("Ingrese los siguientes datos tomando en cuenta las unidades solicitadas")
Qf=float(input("Caudal del fluido en BFPD: "))
BSW=float(input("BWS en %: "))
API=float(input("Gravedad API: "))
ef=float(input("Eficiencia del FWKO en %: "))
ARFWKO=float(input("Ingrese el Aceite Residual (ppm) del equipo: "))
pae=float(input("Ingrese el % de Agua Emulsionada de acuerdo a la tabla NACOT: "))

Qw = (BSW/100)*Qf;
Qo = Qf-Qw;
emulsion = (Qo*100)/(100-pae);
AE = emulsion-Qo;
AL = Qw-AE;

print("FWKO")
# Crear una tabla (DataFrame) vacío con 7 filas y 8 columnas
tabla = pd.DataFrame(index=range(7), columns=range(8))

tabla.iloc[0,0] = "Fluidos"
tabla.iloc[1,0] = "Oil"
tabla.iloc[2,0] = "Water"
tabla.iloc[3,0] = "Emulsión"
tabla.iloc[4,0] = "Agua Libre"
tabla.iloc[5,0] = "Aceite Residual"
tabla.iloc[6,0] = "Total"

# Datos INLET
tabla.iloc[1,1] = Qo
tabla.iloc[2,1] = Qw
tabla.iloc[3,1] = emulsion
tabla.iloc[4,1] = AL
tabla.iloc[5,1] = "-"
tabla.iloc[6,1] = "-"

# Columna 1A
# Eficiencia - Agua Libre
efAL=AL*((100-ef)/100)
Total1A=emulsion+efAL
# Tabla
tabla.iloc[1,2] = Qo
tabla.iloc[2,2] = AE
tabla.iloc[3,2] = emulsion
tabla.iloc[4,2] = efAL
tabla.iloc[5,2] = "-"
tabla.iloc[6,2] = Total1A

# Columna 1B
# Calculos realizados
AL1B=AL-efAL
Total1B=AL1B
# Tabla
tabla.iloc[1,5] = "-"
tabla.iloc[2,5] = "-"
tabla.iloc[3,5] = "-"
tabla.iloc[4,5] = AL1B
tabla.iloc[5,5] = "-"
tabla.iloc[6,5] = Total1B

# Columna 2B - Inicio
AR2B=(AL1B*ARFWKO)/(1000000-ARFWKO)
# Tabla
tabla.iloc[1,6] = "-"
tabla.iloc[3,6] = "-"
tabla.iloc[5,6] = AR2B

# Columna 2A
Qo2A=Qo-AR2B
e2A=(Qo2A*100)/(100-pae)
W2A=e2A-Qo2A
AL2A=efAL
Total2A=e2A+AL2A
# Tabla
tabla.iloc[1,3] = Qo2A
tabla.iloc[2,3] = W2A
tabla.iloc[3,3] = e2A
tabla.iloc[4,3] = AL2A
tabla.iloc[5,3] = "-"
tabla.iloc[6,3] = Total2A

# Columna 2B - Continuación
W2B=AE-W2A
AL2B=AL1B+W2B
Total2B=AL2B+AR2B
# Tabla
tabla.iloc[2,6] = W2B
tabla.iloc[4,6] = AL2B
tabla.iloc[6,6] = Total2B

# Columna 3B - Inicio
AR3B=(AL2B*ARFWKO)/(1000000-ARFWKO)
# Tabla
tabla.iloc[1,7] = "-"
tabla.iloc[3,7] = "-"
tabla.iloc[5,7] = AR3B

# Columna 3A
Qo3A=Qo-AR3B
e3A=(Qo3A*100)/(100-pae)
W3A=e3A-Qo3A
AL3A=efAL
Total3A=e3A+AL3A
# Tabla
tabla.iloc[1,4] = Qo3A
tabla.iloc[2,4] = W3A
tabla.iloc[3,4] = e3A
tabla.iloc[4,4] = AL3A
tabla.iloc[5,4] = "-"
tabla.iloc[6,4] = Total3A

# Columna 3B - Continuación
W3B=W2A-W3A
AL3B=AL2B+W3B
Total3B=AL3B+AR3B
# Tabla
tabla.iloc[2,7] = W3B
tabla.iloc[4,7] = AL3B
tabla.iloc[6,7] = Total3B

#Encabezado
tabla.iloc[0,1] = "Inlet"
tabla.iloc[0,2] = "1 A"
tabla.iloc[0,3] = "2 A"
tabla.iloc[0,4] = "3 A"
tabla.iloc[0,5] = "1 B"
tabla.iloc[0,6] = "2 B"
tabla.iloc[0,7] = "3 B"

# Imprimimos la tabla con las iteraciones

def round_or_keep(value, decimals):
    try:
        return round(float(value), decimals)
    except (ValueError, TypeError):
        return value

# Redondear los valores en la tabla a 4 decimales 
tabla = tabla.applymap(lambda x: round_or_keep(x, 4))
tabla_list = tabla.values.tolist()
print(tabulate(tabla_list, headers=tabla.columns, tablefmt='fancy_grid', showindex=False))

# Hacemos una tabla para analizar el Balance Volumétrico del equipo
TOTAL=Total1A+Total1B
TOTAL2=Total2A+Total2B
TOTAL3=Total3A+Total3B
# Se agregan los datos a la Tabla
tabla1 = pd.DataFrame(index=range(3), columns=range(4))
tabla1.iloc[0,0] = "Total A"
tabla1.iloc[1,0] = "Total B"
tabla1.iloc[2,0] = "Total"
# Total 1A - 1B
tabla1.iloc[0,1] = Total1A
tabla1.iloc[1,1] = Total1B
tabla1.iloc[2,1] = TOTAL
# Total 2A - 2B
tabla1.iloc[0,2] = Total2A
tabla1.iloc[1,2] = Total2B
tabla1.iloc[2,2] = TOTAL2
# Total 3A - 3B
tabla1.iloc[0,3] = Total3A
tabla1.iloc[1,3] = Total3B
tabla1.iloc[2,3] = TOTAL3

# Se imprime la tabla con el Balance Volumetrico
tabla1_list = tabla1.values.tolist()
print(tabulate(tabla1_list, headers=tabla1.columns, tablefmt='fancy_grid'))
```
